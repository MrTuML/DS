```python
df_N = dffinal.loc[dffinal['Nom_Barri'] == 'Baró de Viver']
df_N[['Lloguer mitjà mensual (Euros/mes)','Lloguer mitjà per superfície (Euros/m2 mes)']]
```

Adding the average value for neighborhood in particular case and average value of district for the rest missing values:

```python
# Convertir los valores '--' a NaN en las columnas correspondientes
dffinal['Lloguer mitjà mensual (Euros/mes)'] = pd.to_numeric(dffinal['Lloguer mitjà mensual (Euros/mes)'], errors='coerce')
dffinal['Lloguer mitjà per superfície (Euros/m2 mes)'] = pd.to_numeric(dffinal['Lloguer mitjà per superfície (Euros/m2 mes)'], errors='coerce')

# Calcular la media para el barrio en concreto
barrio = 'Baró de Viver'
df_barrio = dffinal[dffinal['Nom_Barri'] == barrio]
media_mensual = np.nanmean(df_barrio['Lloguer mitjà mensual (Euros/mes)'])
media_superficie = np.nanmean(df_barrio['Lloguer mitjà per superfície (Euros/m2 mes)'])

# Reemplazar los valores NaN por la media del barrio en las columnas correspondientes
dffinal.loc[dffinal['Nom_Barri'] == barrio, 'Lloguer mitjà mensual (Euros/mes)'] = dffinal.loc[df['Nom_Barri'] == barrio, 'Lloguer mitjà mensual (Euros/mes)'].fillna(media_mensual)
dffinal.loc[dffinal['Nom_Barri'] == barrio, 'Lloguer mitjà per superfície (Euros/m2 mes)'] = dffinal.loc[df['Nom_Barri'] == barrio, 'Lloguer mitjà per superfície (Euros/m2 mes)'].fillna(media_superficie)

```

```python
df_N = dffinal.loc[dffinal['Nom_Barri'] == 'Baró de Viver']
df_N[['Lloguer mitjà mensual (Euros/mes)','Lloguer mitjà per superfície (Euros/m2 mes)']]
```

```python
mean_values = dffinal.groupby('Nom_Districte')['Lloguer mitjà mensual (Euros/mes)', 'Lloguer mitjà per superfície (Euros/m2 mes)'].mean()
```

```python
for index, row in dffinal.iterrows():
    districte = row['Nom_Districte']
    if pd.isnull(row['Lloguer mitjà mensual (Euros/mes)']):
        dffinal.loc[index, 'Lloguer mitjà mensual (Euros/mes)'] = mean_values.loc[districte, 'Lloguer mitjà mensual (Euros/mes)']
    if pd.isnull(row['Lloguer mitjà per superfície (Euros/m2 mes)']):
        dffinal.loc[index, 'Lloguer mitjà per superfície (Euros/m2 mes)'] = mean_values.loc[districte, 'Lloguer mitjà per superfície (Euros/m2 mes)']
```

```python
dffinal.loc[(dffinal['Lloguer mitjà mensual (Euros/mes)'] == '--') & (dffinal['Lloguer mitjà per superfície (Euros/m2 mes)'] == '--')].value_counts()
```

Now data is ready to plot

###Visualization of a categorical variable (District or Neighborhood):

```python
plt.figure(figsize=(15, 10))
district_counts = dffinal['Nom_Districte'].value_counts()
neighborhood_counts = df.groupby('Nom_Districte')['Nom_Barri'].nunique()

district_counts.plot(kind='bar')
plt.xlabel('District')
plt.ylabel('Frequency')
plt.title('Distribution of Rentals by District')

for i, count in enumerate(district_counts):
    district = district_counts.index[i]
    neighborhood_count = neighborhood_counts[district]
    plt.text(i, count, f"Nhoods: {neighborhood_count}", ha='center', va='bottom')

plt.tight_layout()  # Ajustar los subplots para evitar solapamientos

plt.savefig('rentals_distribution.png', dpi=300)  # Guardar la figura como imagen PNG

plt.show()
```

As expected, the district with the highest rental rate is the one that has the most neighborhoods.

### A numeric variable (Price)

```python
plt.figure(figsize=(10, 6))

# Histograma para Lloguer mitjà mensual
plt.subplot(1, 2, 1)  # Subplot izquierdo
sns.histplot(dffinal['Lloguer mitjà mensual (Euros/mes)'], kde=True)
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Distribution of Rental Prices')

# Histograma para Lloguer mitjà per superfície
plt.subplot(1, 2, 2)  # Subplot derecho
sns.histplot(dffinal['Lloguer mitjà per superfície (Euros/m2 mes)'], kde=True)
plt.xlabel('Price per Surface')
plt.ylabel('Frequency')
plt.title('Distribution of Rental Prices per Surface')

plt.tight_layout()  # Ajustar los subplots para evitar solapamientos
# Guardar el gráfico como imagen PNG
plt.savefig('histogramprice.png', dpi=300)
plt.show()
```

###A numerical and a categorical variable (District and Price)

```python
plt.figure(figsize=(15, 10))
plt.subplot(2, 1, 1) # Subplot 1
sns.barplot(x='Nom_Districte', y='Lloguer mitjà mensual (Euros/mes)', data=dffinal, color='blue')
plt.xlabel('District')
plt.ylabel('Monthly Rent (Euros)')
plt.title('Average Monthly Rent by District')

plt.savefig('monthly_rent_by_district.png') # Guardar el gráfico como imagen

# Crear el segundo gráfico - Alquiler por metro cuadrado promedio por distrito
plt.figure(figsize=(15, 10))
plt.subplot(2, 1, 2) # Subplot 2
sns.barplot(x='Nom_Districte', y='Lloguer mitjà per superfície (Euros/m2 mes)', data=dffinal, color='orange')
plt.xlabel('District')
plt.ylabel('Rent per Square Meter (Euros/m2)')
plt.title('Average Rent per Square Meter by District')

plt.savefig('rent_per_square_meter_by_district.png') # Guardar el gráfico como imagen
```

```python
plt.figure(figsize=(15, 15))
plt.subplot(2, 1, 1)
sns.boxplot(x='Nom_Districte', y='Lloguer mitjà mensual (Euros/mes)', data=dffinal)
plt.xlabel('District')
plt.ylabel('Monthly Rent (Euros/mes)')
plt.title('Rental Prices by District')

plt.subplot(2, 1, 2)
sns.boxplot(x='Nom_Districte', y='Lloguer mitjà per superfície (Euros/m2 mes)', data=dffinal)
plt.xlabel('District')
plt.ylabel('Rent per Surface (Euros/m2 mes)')
plt.title('Rental Prices per Surface by District')

plt.xticks(rotation=90)
plt.tight_layout()
plt.savefig('boxplot_rent.png')  # Guardar la imagen como PNG
plt.show()
```

###Two numeric variables (Year or Quarter and Price)

```python
plt.figure(figsize=(10, 6))

# Plot for Lloguer mitjà mensual
plt.subplot(2, 1, 1)
sns.lineplot(x='Trimestre', y='Lloguer mitjà mensual (Euros/mes)', data=dffinal)
plt.xlabel('Trimestre')
plt.ylabel('Monthly Rent (Euros/mes)')
plt.title('Monthly Rent Trend by Trimestre')

# Plot for Lloguer mitjà per superfície
plt.subplot(2, 1, 2)
sns.lineplot(x='Trimestre', y='Lloguer mitjà per superfície (Euros/m2 mes)', data=dffinal)
plt.xlabel('Trimestre')
plt.ylabel('Rent per Surface (Euros/m2 mes)')
plt.title('Rent per Surface Trend by Trimestre')

plt.tight_layout()
plt.savefig('lineplot_price_trimestre.png')
plt.show()
```

### Three variables (Neighborhood or District, Quarter and Price)

```python
fig, axes = plt.subplots(4, 1, figsize=(10, 20))

# Gráfico 1: Tres variables (Nom_Barri, Trimestre, Lloguer mitjà mensual (Euros/mes))
sns.barplot(ax=axes[0], x='Nom_Barri', y='Lloguer mitjà mensual (Euros/mes)', hue='Trimestre', data=dffinal)
axes[0].set_xlabel('Nom_Barri')
axes[0].set_ylabel('Monthly Rent (Euros/mes)')
axes[0].set_title('Rent by Nom_Barri, Trimestre, and Monthly Rent')
axes[0].legend(title='Trimestre')
axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=90, ha='center')

# Gráfico 2: Tres variables (Nom_Barri, Trimestre, Lloguer mitjà per superfície (Euros/m2 mes))
sns.barplot(ax=axes[1], x='Nom_Barri', y='Lloguer mitjà per superfície (Euros/m2 mes)', hue='Trimestre', data=dffinal)
axes[1].set_xlabel('Nom_Barri')
axes[1].set_ylabel('Rent per Surface (Euros/m2 mes)')
axes[1].set_title('Rent by Nom_Barri, Trimestre, and Rent per Surface')
axes[1].legend(title='Trimestre')
axes[1].set_xticklabels(axes[1].get_xticklabels(), rotation=90, ha='center')

# Gráfico 3: Tres variables (Nom_Districte, Trimestre, Lloguer mitjà mensual (Euros/mes))
sns.barplot(ax=axes[2], x='Nom_Districte', y='Lloguer mitjà mensual (Euros/mes)', hue='Trimestre', data=dffinal)
axes[2].set_xlabel('Nom_Districte')
axes[2].set_ylabel('Monthly Rent (Euros/mes)')
axes[2].set_title('Rent by Nom_Districte, Trimestre, and Monthly Rent')
axes[2].legend(title='Trimestre')
axes[2].set_xticklabels(axes[2].get_xticklabels(), rotation=90, ha='center')

# Gráfico 4: Tres variables (Nom_Districte, Trimestre, Lloguer mitjà per superfície (Euros/m2 mes))
sns.barplot(ax=axes[3], x='Nom_Districte', y='Lloguer mitjà per superfície (Euros/m2 mes)', hue='Trimestre', data=dffinal)
axes[3].set_xlabel('Nom_Districte')
axes[3].set_ylabel('Rent per Surface (Euros/m2 mes)')
axes[3].set_title('Rent by Nom_Districte, Trimestre, and Rent per Surface')
axes[3].legend(title='Trimestre')
axes[3].set_xticklabels(axes[3].get_xticklabels(), rotation=90, ha='center')

plt.tight_layout()
plt.savefig('four_barplots.png')
plt.show()
```

```python
fig, axes = plt.subplots(4, 1, figsize=(10, 24))

# Violin plot para "Nom_Barri", "Trimestre" y "Lloguer mitjà mensual (Euros/mes)"
sns.violinplot(ax=axes[0], x='Nom_Barri', y='Lloguer mitjà mensual (Euros/mes)', hue='Trimestre', data=dffinal)
axes[0].set_xlabel('Barri')
axes[0].set_ylabel('Monthly Rent (Euros/mes)')
axes[0].set_title('Rent by Barri and Trimestre')
axes[0].legend(title='Trimestre')
axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=90)

# Violin plot para "Nom_Barri", "Trimestre" y "Lloguer mitjà per superfície (Euros/m2 mes)"
sns.violinplot(ax=axes[1], x='Nom_Barri', y='Lloguer mitjà per superfície (Euros/m2 mes)', hue='Trimestre', data=dffinal)
axes[1].set_xlabel('Barri')
axes[1].set_ylabel('Rent per Surface (Euros/m2 mes)')
axes[1].set_title('Rent per Surface by Barri and Trimestre')
axes[1].legend(title='Trimestre')
axes[1].set_xticklabels(axes[1].get_xticklabels(), rotation=90)

# Violin plot para "Nom_Districte", "Trimestre" y "Lloguer mitjà mensual (Euros/mes)"
sns.violinplot(ax=axes[2], x='Nom_Districte', y='Lloguer mitjà mensual (Euros/mes)', hue='Trimestre', data=dffinal)
axes[2].set_xlabel('District')
axes[2].set_ylabel('Monthly Rent (Euros/mes)')
axes[2].set_title('Rent by District and Trimestre')
axes[2].legend(title='Trimestre')
axes[2].set_xticklabels(axes[2].get_xticklabels(), rotation=90)

# Violin plot para "Nom_Districte", "Trimestre" y "Lloguer mitjà per superfície (Euros/m2 mes)"
sns.violinplot(ax=axes[3], x='Nom_Districte', y='Lloguer mitjà per superfície (Euros/m2 mes)', hue='Trimestre', data=dffinal)
axes[3].set_xlabel('District')
axes[3].set_ylabel('Rent per Surface (Euros/m2 mes)')
axes[3].set_title('Rent per Surface by District and Trimestre')
axes[3].legend(title='Trimestre')
axes[3].set_xticklabels(axes[3].get_xticklabels(), rotation=90)

plt.tight_layout()
plt.savefig("violin_plot.png")
plt.show()
```
