```python
import nltk
from nltk.tokenize import sent_tokenize
```

```python
nltk.download('punkt')
```

```python
tokenized_text=sent_tokenize(tlptext)
print(tokenized_text)
```

```python
# split into words
from nltk.tokenize import word_tokenize
tokens = word_tokenize(tlptext)
print(tokens[:100])
```

```python
# convert to lower case
tokens = [w.lower() for w in tokens]
```

```python
# remove all tokens that are not alphabetic
words = [word for word in tokens if word.isalpha()]
print(words[:100])
```

```python
from nltk.probability import FreqDist
freqtlp = FreqDist(words)
print(freqtlp)
```

```python
# Frequency Distribution Plot
import matplotlib.pyplot as plt
plt.figure(figsize=(16, 8))
freqtlp.plot(100,cumulative=False)
plt.show()
```

```python
# filter out stop words
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
stwords = [w for w in words if not w in stop_words]
print(stwords[:100])
```

```python
freqtlpsw = FreqDist(stwords)
print(freqtlp)
# Frequency Distribution Plot
import matplotlib.pyplot as plt
plt.figure(figsize=(16, 8))
freqtlpsw.plot(100,cumulative=False)
plt.show()
```

```python
# stemming of words
from nltk.stem.porter import PorterStemmer
porter = PorterStemmer()
stemmed = [porter.stem(word) for word in stwords]
print(stemmed[:100])
```

Així, a primera vista es pot observar com, per exemple, la paraula "little" la transforma en "littl" i "prince" en "princ", entre d'altres. La primera no existeix dins del diccionari anglès i la segona té un significat diferent a la paraula original. Per tant, l'algorisme de stemming de Porter en aquest cas no ofereix gaire precisió.

-----------------------------------------------------------------

Así,a primera vista se observa cómo, por ejemplo, la palabra little la transforma a "littl" y prince a "princ" entre otras. La primera no existe dentro del diccionario inglés y la segunda tiene un significado diferente a la palabra original. So, at first glance, it can be observed how, for example, the word "little" is transformed into "littl" and "prince" into "princ," among others.
```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()

# Lematizar las palabras
lemmatized = [lemmatizer.lemmatize(word) for word in stwords]

print(lemmatized[:100])
```

```python
freqtllmw = FreqDist(lemmatized)
print(freqtllmw)
# Frequency Distribution Plot
plt.figure(figsize=(16, 8))
freqtllmw.plot(100,cumulative=False)
plt.show()
```
