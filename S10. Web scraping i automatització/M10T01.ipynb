```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL de la página web
url = 'https://www.bolsasymercados.es/bme-exchange/en/Prices-and-Markets/Shares/Main-Market/Prices/ibex-35-ES0SI0000005'

# Realiza una solicitud HTTP GET para obtener el contenido de la página
response = requests.get(url)

# Comprueba si la solicitud fue exitosa (código de respuesta 200)
if response.status_code == 200:
    # Parsea el contenido HTML con BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Encuentra la tabla por clase CSS
    table = soup.find('table', class_='shares-table')

    # Extrae las filas de la tabla
    rows = table.find_all('tr')

    # Inicializa listas para cada columna de datos
    names = []
    last_prices = []
    percent_diffs = []
    highs = []
    lows = []
    volumes = []
    turnovers = []
    dates = []
    times = []

    # Recorre las filas de la tabla y extrae los datos de cada columna
    for row in rows[1:]:  # Comenzamos desde la segunda fila para omitir la fila de encabezado
        columns = row.find_all('td')
        names.append(columns[0].text.strip())
        last_prices.append(columns[1].text.strip())
        percent_diffs.append(columns[2].text.strip())
        highs.append(columns[3].text.strip())
        lows.append(columns[4].text.strip())
        volumes.append(columns[5].text.strip().replace(',', ''))  # Elimina las comas de los números
        turnovers.append(columns[6].text.strip().replace(',', ''))  # Elimina las comas de los números
        dates.append(columns[7].text.strip())
        times.append(columns[8].text.strip())

    # Crea un DataFrame con los datos extraídos
    data = {
        'Name': names,
        'Last': last_prices,
        '% Diff.': percent_diffs,
        'High': highs,
        'Low': lows,
        'Volume': volumes,
        'Turnover (€ Thousands)': turnovers,
        'Date': dates,
        'Time': times
    }

    df = pd.DataFrame(data)

    # Imprime el DataFrame
    print(df)

else:
    print(f'Error al obtener la página web (código de respuesta: {response.status_code})')
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import pandas as pd

# Configura el servicio del driver de Chrome
service = Service(executable_path=r'/usr/bin/chromedriver')

# Configura las opciones de Chrome
options = Options()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')

# Inicializa el navegador Chrome con la configuración
driver = webdriver.Chrome(service=service, options=options)
driver.get('https://quotes.toscrape.com/')

# Encuentra el elemento que contiene la cita
quote_element = driver.find_element(By.CSS_SELECTOR, 'div.quote span.text')

# Encuentra el elemento que contiene el nombre del autor
author_element = driver.find_element(By.CSS_SELECTOR, 'div.quote span small.author')

# Extrae el texto de la cita y el nombre del autor
quote_text = quote_element.text
author_name = author_element.text

# Imprime el resultado
print(f'Autor: {author_name}')
print(f'Cita: {quote_text}')

# Cierra el navegador
driver.quit()


# Crea un DataFrame con los datos
data = {'Autor': [author_name], 'Cita': [quote_text]}
df = pd.DataFrame(data)

# Guarda el DataFrame en un archivo CSV
df.to_csv('citassl.csv', index=False)
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import pandas as pd

# Configura el servicio de ChromeDriver
service = Service(executable_path='/usr/bin/chromedriver')

# Configura las opciones de Chrome
options = Options()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')

# Inicializa el navegador web
with webdriver.Chrome(service=service, options=options) as driver:
    # Abre la página web con la tabla de datos
    url = 'https://www.bolsasymercados.es/bme-exchange/en/Prices-and-Markets/Shares/Main-Market/Prices/ibex-35-ES0SI0000005'
    driver.get(url)

    # Espera a que la página cargue completamente (puedes ajustar el tiempo según tu conexión)
    driver.implicitly_wait(20)

    # Encuentra la tabla
    my_table = driver.find_element('xpath', '/html/body/main/div/div/div/div[2]/div[5]/table')

    # Encuentra los nombres de las columnas
    names = my_table.find_elements('xpath', '//thead/tr/th')
    column_names = [name.text for name in names]
    column_names = column_names[-9:]
    # Encuentra los datos de la tabla
    data = []
    rows = my_table.find_elements('xpath', '//tbody/tr')
    for row in rows:
        row_data = [cell.text for cell in row.find_elements('tag name', 'td')]
        data.append(row_data)

    # Crea un DataFrame con los datos y los nombres de las columnas
    df = pd.DataFrame(data, columns=column_names)

    # Imprime el DataFrame
    print(df)
```
