```python
dff["AvgSpeed"] = dff["Distance"] / (dff["AirTime"]/60)
#dff.head()
dff.tail()
```

```python
# Columna indicando si el vuelo llegó tarde o no (1 si llegó tarde, 0 si llegó a tiempo)
dff["LateArrival"] = (dff["ArrDelay"] > 0).astype(int)
dff.head()
```

```python
# Aerolíneas con mayores retrasos acumulados:
airline_delays = dff.groupby("UniqueCarrier")["ArrDelay"].sum().sort_values(ascending=False)
print(airline_delays)
```

```python
# Vuelos más largos (mayor distancia)
longest_flights = dff.nlargest(10, "Distance")
print(longest_flights[["FlightNum", "Origin", "Dest", "Distance"]])
```

```python
# Vuelos más atrasados (mayor retraso en la llegada)
most_delayed_flights = dff.nlargest(10, "ArrDelay")
print(most_delayed_flights[["FlightNum", "Origin", "Dest", "ArrDelay"]])
```
